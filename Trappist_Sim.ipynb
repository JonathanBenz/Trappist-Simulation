```python
# Importing all necessary packages
import rebound
import matplotlib.pyplot as plt
import numpy as np

# Setting up the simulation and units
sim = rebound.Simulation()
sim.units = ['Msun', 'yr', 'AU']
sim.G = 4*np.pi**2
```

```python
# Dataset 1
# Mass in [solar masses], a in [AU]
# Data taken from Gillon et al., 2017
sim.add(m=0.089)                             # Trappist-1a
sim.add(m=3.06278e-6, a=0.01150,  e=0.081)   # Trappist-1b
sim.add(m=3.48317e-6, a=0.01576,  e=0.083)   # Trappist-1c
sim.add(m=8.918108e-7, a=0.02219,  e=0.070)  # Trappist-1d
sim.add(m=2.318108e-6, a=0.02916,  e=0.085)  # Trappist-1e
sim.add(m=2.80455e-6, a=0.03836,  e=0.063)   # Trappist-1f
sim.add(m=3.447134e-6, a=0.0467,  e=0.061)   # Trappist-1g
sim.add(m=9.939037e-7, a=0.0617)             # Trappist-1h
sim.move_to_com()
```

```python
# Dataset 2
# Mass in [solar masses], a in [AU]
# Data taken from Grimm et al., 2018
sim.add(m=0.089)                                 # Trappist-1a
sim.add(m=4.1350597e-6, a=0.01154,  e=0.00622)   # Trappist-1b
sim.add(m=3.9350778e-6, a=0.01580,  e=0.00654)   # Trappist-1c
sim.add(m=1.1665607e-6, a=0.02227,  e=0.00837)   # Trappist-1d
sim.add(m=2.0814925e-6, a=0.02925,  e=0.00510)   # Trappist-1e
sim.add(m=3.1261423e-6, a=0.03849,  e=0.01007)   # Trappist-1f
sim.add(m=3.9750141e-6, a=0.04683,  e=0.00208)   # Trappist-1g
sim.add(m=9.7919029e-7, a=0.06189, e=0.00567)    # Trappist-1h
sim.move_to_com()
```

```python
# Storing each value of x_pos, y_pos, eccentricity, and semi_major_axis from iterating through 1000 years
timestep = 1000
x_pos = np.empty((8,timestep))
y_pos = np.empty((8,timestep))
eccentricity = np.empty((8,timestep))
semi_major_axis = np.empty((8,timestep))
times = np.linspace(0, 1000, timestep)
for i,t in enumerate(times):
  sim.integrate(t)
  x_pos[1][i] = sim.particles[1].x
  y_pos[1][i] = sim.particles[1].y
  eccentricity[1][i] = sim.particles[1].e
  semi_major_axis[1][i] = sim.particles[1].a
  x_pos[2][i] = sim.particles[2].x
  y_pos[2][i] = sim.particles[2].y
  eccentricity[2][i] = sim.particles[2].e
  semi_major_axis[2][i] = sim.particles[2].a
  x_pos[3][i] = sim.particles[3].x
  y_pos[3][i] = sim.particles[3].y
  eccentricity[3][i] = sim.particles[3].e
  semi_major_axis[3][i] = sim.particles[3].a
  x_pos[4][i] = sim.particles[4].x
  y_pos[4][i] = sim.particles[4].y
  eccentricity[4][i] = sim.particles[4].e
  semi_major_axis[4][i] = sim.particles[4].a
  x_pos[5][i] = sim.particles[5].x
  y_pos[5][i] = sim.particles[5].y
  eccentricity[5][i] = sim.particles[5].e
  semi_major_axis[5][i] = sim.particles[5].a
  x_pos[6][i] = sim.particles[6].x
  y_pos[6][i] = sim.particles[6].y
  eccentricity[6][i] = sim.particles[6].e
  semi_major_axis[6][i] = sim.particles[6].a
  x_pos[7][i] = sim.particles[7].x
  y_pos[7][i] = sim.particles[7].y
  eccentricity[7][i] = sim.particles[7].e
  semi_major_axis[7][i] = sim.particles[7].a
```

```python
# Plotting the x and y positions
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect("equal")

# manually set plot boundaries
lim = 0.065
ax.set_xlim([-lim, lim])
ax.set_ylim([-lim, lim])

plt.scatter(x_pos[0], y_pos[0], s=50, marker='*', facecolor='yellow', zorder=3)
plt.scatter(x_pos[1], y_pos[1], s=5, facecolor='darkslateblue', zorder=3)
plt.scatter(x_pos[2], y_pos[2], s=5, facecolor='royalblue', zorder=3)
plt.scatter(x_pos[3], y_pos[3], s=5, facecolor='cyan', zorder=3)
plt.scatter(x_pos[4], y_pos[4], s=5, facecolor='mediumseagreen', zorder=3)
plt.scatter(x_pos[5], y_pos[5], s=5, facecolor='palegreen', zorder=3)
plt.scatter(x_pos[6], y_pos[6], s=5, facecolor='violet', zorder=3)
plt.scatter(x_pos[7], y_pos[7], s=5, facecolor='deeppink', zorder=3)
plt.xlabel('x [AU]')
plt.ylabel('y [AU]')
plt.legend(['1a', '1b', '1c', '1d', '1e', '1f', '1g', '1h'])
plt.show()
```

```python
# Eccentricity vs Time Plot
plt.plot(times, eccentricity[1],'darkslateblue')
plt.plot(times, eccentricity[2],'royalblue')
plt.plot(times, eccentricity[3],'cyan')
plt.plot(times, eccentricity[4],'mediumseagreen')
plt.plot(times, eccentricity[5],'palegreen')
plt.plot(times, eccentricity[6],'violet')
plt.plot(times, eccentricity[7],'deeppink')
one = np.ones_like((eccentricity[1]))
plt.plot(times, one, 'k')
plt.title('Eccentricity vs Time')
plt.xlabel('time [yrs]')
plt.ylabel('eccentricity')
plt.legend(['1b', '1c', '1d', '1e', '1f', '1g', '1h', 'e=1 Threshold'])

# For e < 1, orbit is elliptical. For e = 1, orbit is parabolic. For e > 1, orbit is hyperbolic. 
```

```python
# Semimajor axis vs Time Plot
plt.plot(times, semi_major_axis[1],'darkslateblue')
plt.plot(times, semi_major_axis[2],'royalblue')
plt.plot(times, semi_major_axis[3],'cyan')
plt.plot(times, semi_major_axis[4],'mediumseagreen')
plt.plot(times, semi_major_axis[5],'palegreen')
plt.plot(times, semi_major_axis[6],'violet')
plt.plot(times, semi_major_axis[7],'deeppink')
plt.title('Semi-major Axis vs Time')
plt.xlabel('time [yrs]')
plt.ylabel('semi-major axis [AU]')
plt.legend(['1b', '1c', '1d', '1e', '1f', '1g', '1h'])
```
